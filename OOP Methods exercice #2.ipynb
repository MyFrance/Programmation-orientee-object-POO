```python
class Product:
    def __init__(self, productName, price):
        self.productName = productName
        self.price = price 
        
    def discount(self, discountValue):
        discountAmount = (self.price * discountValue)/100
        finalPrice = self.price - discountAmount
        return finalPrice
```

```python
laptop = Product("laptop", 1552)
laptop.discount(10)
```

```python
mobil = Product("mobil", 1900)
laptop.discount(15)
```
