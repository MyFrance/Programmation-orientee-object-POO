```python
import math
```

```python
class Cercle:
    pie = math.pi
    def __init__(self, rayon): 
        self.rayon = rayon
    
        
    def calcul_perimetre(self): 
        formule_perimetre = 2 * Cercle.pie * self.rayon
        return formule_perimetre
    
    def calcul_air(self): 
        formule_air = Cercle.pie*self.rayon**2
        return formule_air
```

```python
cercle = Cercle(24)
print(f"Le périmètre du cercle est : {round(cercle.calcul_perimetre(), 2)}")
print(f"L'air du cercle est : {round(cercle.calcul_air(), 2)}")
```
