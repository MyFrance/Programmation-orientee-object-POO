```python
class MobilPhone():
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def discount(self, discountValue):
        discountAmount = (self.price * discountValue)/100
        discountPrice = self.price - discountAmount
        return discountPrice
```

```python
galaxyS9 = MobilPhone("galaxyS9", 800)
```
